```python
import pandas as pd

# Read Excel file
df = pd.read_csv("data/2018-v269-16052023-EU MRV Publication of information.csv")

# Display data
df
```

```python
# Display all the columns in the dataframe
df.columns
```

```python
# Print the number of ships per ship type
countByType = df.groupby('Ship type')['Name'].nunique()
print(countByType)
```

```python
# Sum of fuel consumption by ship type
fuelConsumptionByShipType = df.groupby('Ship type')['Total fuel consumption [m tonnes]'].sum()
fuelConsumptionByShipType
```

```python
# Sum of CO2 emmision per ship type
co2EmissionPerShipType = df.groupby('Ship type')['Total CO₂ emissions [m tonnes]'].sum()
co2EmissionPerShipType
```

```python
# Total steaming time

steamingTimeByShipType = df.groupby('Ship type')['Total time spent at sea [hours]'].sum()
steamingTimeByShipType
```

```python
# Print the number of ships per verifier
countByType = df.groupby('Verifier Name')['Name'].nunique()
print(countByType)
```

```python
# Count of vessels grouped by vessel type under CII Rating categories


# Create a new DataFrame to store the counts
vessel_counts = pd.DataFrame(columns=['vessel_type', 'category', 'count'])

# Iterate through the CII Rating categories
for category in ['A', 'B', 'C', 'D']:
    category_counts = df[df[category] == 'Yes'].groupby('Ship type').size().reset_index()
    category_counts.columns = ['vessel_type', 'count']
    category_counts['category'] = category
    # Add the counts to the vessel_counts DataFrame
    vessel_counts = pd.concat([vessel_counts, category_counts])
    # print(result)
    
# Sort the DataFrame by vessel type
vessel_counts = vessel_counts.sort_values('vessel_type')
vessel_counts

```
