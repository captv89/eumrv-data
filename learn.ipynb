```python
import pandas as pd

# Read Excel file
df = pd.read_csv("data/2018-v269-16052023-EU MRV Publication of information.csv")

# Display data
df
```

```python
# Display all the columns in the dataframe
df.columns
```

```python
# Print the number of ships per ship type
countByType = df.groupby('Ship type')['Name'].nunique()
print(countByType)
```

```python
# Sum of fuel consumption by ship type
fuelConsumptionByShipType = df.groupby('Ship type')['Total fuel consumption [m tonnes]'].sum()
fuelConsumptionByShipType
```

```python
# Sum of CO2 emmision per ship type
co2EmissionPerShipType = df.groupby('Ship type')['Total CO₂ emissions [m tonnes]'].sum()
co2EmissionPerShipType
```

```python
# Total steaming time

steamingTimeByShipType = df.groupby('Ship type')['Total time spent at sea [hours]'].sum()
steamingTimeByShipType
```

```python
# Print the number of ships per verifier
countByType = df.groupby('Verifier Name')['Name'].nunique()
print(countByType)
```

```python
# Count of vessels grouped by vessel type under CII Rating categories


# Create a new DataFrame to store the counts
vessel_counts = pd.DataFrame(columns=['vessel_type', 'category', 'count'])

# Iterate through the CII Rating categories
for category in ['A', 'B', 'C', 'D']:
    category_counts = df[df[category] == 'Yes'].groupby('Ship type').size().reset_index()
    category_counts.columns = ['vessel_type', 'count']
    category_counts['category'] = category
    # Add the counts to the vessel_counts DataFrame
    vessel_counts = pd.concat([vessel_counts, category_counts])
    # print(result)
    
# Sort the DataFrame by vessel type
vessel_counts = vessel_counts.sort_values('vessel_type')
vessel_counts

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# year and filepath
file_path = {
    '2018': 'data/2018-v269-16052023-EU MRV Publication of information.csv',
    '2019': 'data/2019-v215-30062023-EU MRV Publication of information.csv',
    '2020': 'data/2020-v191-20072023-EU MRV Publication of information.csv',
    '2021': 'data/2021-v169-02082023-EU MRV Publication of information.csv',
    '2022': 'data/2022-v35-05082023-EU MRV Publication of information.csv',
}

# Combine all the data into one DataFrame
all_data = pd.DataFrame()
for year in file_path:
    print(year)
    data = pd.read_csv(file_path[year])
    data['Year'] = year
    all_data = pd.concat([all_data, data])


# Preprocessing
all_data['Reporting Period'] = pd.to_datetime(all_data['Reporting Period'], format='%Y')

# Print the Reporting Period range
print("Reporting Period range: ", all_data['Reporting Period'].min(), " to ", all_data['Reporting Period'].max())
# all_data.set_index('Reporting Period', inplace=True)

# EDA - Summary statistics
summary_stats = all_data.describe()

print(summary_stats)
```

```python
# EDA - Visualizations
plt.figure(figsize=(10, 6))
sns.histplot(all_data['Total CO₂ emissions [m tonnes]'], bins=20, kde=True)
plt.title('Distribution of Total CO₂ Emissions')
plt.xlabel('CO₂ Emissions (m tonnes)')
plt.ylabel('Frequency')
plt.show()

```

```python
data = all_data.copy()

# Print the Reporting Period range
print("Reporting Period range: ", data['Reporting Period'].min(), " to ", data['Reporting Period'].max())


# Clean the 'Annual average Fuel consumption per distance [kg / n mile]' column
data['Annual average Fuel consumption per distance [kg / n mile]'] = pd.to_numeric(data['Annual average Fuel consumption per distance [kg / n mile]'], errors='coerce')


# Identify vessels that have improved over the last 5 years
data['4-year change'] = data.groupby('IMO Number')['Annual average Fuel consumption per distance [kg / n mile]'].diff(periods=4)
improved_vessels = data[data['4-year change'] < 0]

# print("Vessels that have improved over the last 4 years:")


# Sort improved vessels from most changed to least
sorted_improved_vessels = improved_vessels.sort_values(by='4-year change', ascending=False)

# Get the top 10 vessels with the most improvement
top_improved_vessels = sorted_improved_vessels.nlargest(10, '4-year change')
# print(top_improved_vessels)


# Filter data for the last 3 years and selected vessels
filtered_data = data[data['IMO Number'].isin(top_improved_vessels['IMO Number'])]
filtered_data = filtered_data[filtered_data['Reporting Period'].dt.year >= (filtered_data['Reporting Period'].max().year - 3)]
# print(filtered_data)


# Create a pivot table
pivot_table = filtered_data.pivot_table(values='Annual average Fuel consumption per distance [kg / n mile]',
                                        index='Reporting Period', columns='Name', aggfunc='first')

print(pivot_table)

# Plot the data
pivot_table.plot(marker='o', figsize=(10, 6))
plt.title("Annual average Fuel consumption per distance [kg / n mile] for Top 10 Improved Vessels")
plt.xlabel("Reporting Period")
plt.ylabel("Annual average Fuel consumption per distance [kg / n mile]")
plt.legend(title="IMO Number")
plt.grid(True)
plt.show()



```
